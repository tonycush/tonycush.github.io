### Import my libraries & load excel doc

```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np

company_data = pd.read_excel('../my_files/features.xlsx')
sic_list = pd.read_csv('../my_files/SIC list.csv')
```

```python
company_data
```

```python
industries = []
for col in company_data.columns:
    if col[:4] == "ind_":
        industries.append(col)

industries_tots = []
for industry in industries:
    total = company_data[industry].sum()
    industries_tots.append(total)

industry_bars = plt.bar(industries,industries_tots)

plt.show()
```

```python
# only plotting popular industries
pop_industries_tots = []
pop_industries = []
for industry in industries:
    total = company_data[industry].sum()
    if total > 25:
        pop_industries.append(industry[4:])
        pop_industries_tots.append(total)
     
plt.figure(figsize=(5,3),dpi=150)
plt.xticks(rotation=45, ha='right')
pop_industry_bars = plt.bar(pop_industries,pop_industries_tots)

plt.title('Popular Industries', fontdict={'fontname':'Arial', 'fontsize': 16})  

plt.show()
```


## SIC
A graph on the Industry Coding

```python
sics = []
for col in company_data.columns:
    if col[:4] == "sic_":
        sics.append(col)

sics_tots = []
for sic in sics:
    total = company_data[sic].sum()
    sics_tots.append(total)

industry_bars = plt.bar(sics,sics_tots)

plt.show()
```

```python
# only plotting popular sics
def find_sic_name(code,sic_list):
    sic_index = sic_list[sic_list['SIC Code']==int(code)].index.values
    name = (sic_list['Description'].iloc[sic_index].values)
    return name[0]

pop_sics_tots = []
pop_sics = []
for sic in sics:
    total = company_data[sic].sum()
    if total > 10:
        name = find_sic_name(sic[4:],sic_list)
        pop_sics.append(name)
        pop_sics_tots.append(total)


plt.figure(figsize=(5,3),dpi=150)
plt.xticks(rotation=-30, ha='left')
pop_sics_bars = plt.bar(pop_sics,pop_sics_tots)

plt.title('Popular SICs', fontdict={'fontname':'Arial', 'fontsize': 16})  
#plt.xticks.setrotation(55)


plt.show()
```

### INVESTORS

```python
investors = []
for col in company_data.columns:
    if col[:5] == "invr_":
        investors.append(col)

investor_tots = []
for investor in investors:
    total = company_data[investor].sum()
    investor_tots.append(total)

industry_bars = plt.bar(investors,investor_tots)

plt.show()
```

```python
# only plotting popular industries
pop_investors_tots = []
pop_investors = []
for investor in investors:
    total = company_data[investor].sum()
    if total > 4:
        pop_investors.append(investor[5:])
        pop_investors_tots.append(total)
     
plt.figure(figsize=(5,3),dpi=150)
plt.xticks(rotation=45, ha='right')
pop_industry_bars = plt.bar(pop_investors,pop_investors_tots)

plt.title('TOP INVESTORS', fontdict={'fontname':'Arial', 'fontsize': 16}) 
plt.ylabel('No. of Companies', fontdict={'fontname':'Arial', 'fontsize': 12}) 

plt.show()
```

### FUNDING ROUND TYPES

```python
funding_rounds = []
for col in company_data.columns:
    if col[:3] == "fr_":
        funding_rounds.append(col)
        
funding_round_names = []
funding_round_tots = []
for funding_round in funding_rounds:
    total = company_data[funding_round].sum()
    funding_round_tots.append(total)
    funding_round_names.append(funding_round[3:])

plt.figure(figsize=(5,3),dpi=150)
plt.xticks(rotation=45, ha='right')
funding_bars = plt.bar(funding_round_names,funding_round_tots)
plt.title('FUNDING ROUNDS', fontdict={'fontname':'Arial', 'fontsize': 16})  

plt.show()
```

```python
ipo_status = company_data['IPO Status ']
ipo_status_type = ipo_status.unique()
ipo_status_count = []

for ipo in ipo_status_type:
    this_ipo = ipo_status[ipo_status==ipo]
    ipo_count = this_ipo.count()
    ipo_status_count.append(ipo_count)
    
plt.figure(figsize=(5,3),dpi=150)
#plt.xticks(rotation=45, ha='right')
ipo_bars = plt.bar(ipo_status_type,ipo_status_count)

plt.title('IPO Status', fontdict={'fontname':'Arial', 'fontsize': 16})  

plt.show()
```

```python

distinct_titles = []
for i in range(12):
    title = 'y'+str(i)+'_title'
    year_headers = []
    for col in company_data.columns:
        if col[:8] == title or col[:9]==title:
            year_headers.append(col)
            deets = col.split('_')
            if deets[2] not in distinct_titles:
                distinct_titles.append(deets[2])
    print('year '+str(i)+' has '+str(len(year_headers))+' headers')
print(len(distinct_titles))
for title in distinct_titles:
    print(title)
```

```python
year0 = sub_company_data[y0_acc_headers]
year0= year0.set_index('name')
year0 = year0.T
year0 = year0.dropna(how='all')
print(year0.info())
print(year0)
```

```python
def transpose_append_annual(df,headers,full_df):
    this_year= df[headers]
    #removing the prefix from column titles
    #this_year_cols = [header.split('_')[2]+'_'+header.split('_')[0] for header in headers[1:]]
    #this_year_cols.insert(0,'name')
    #this_year.columns = this_year_cols
    this_year = this_year.set_index('name')
    this_year = this_year.T
    #this_year = this_year.dropna(how='all')
    full_df = pd.concat([full_df,this_year])
    return full_df
    
all_accounts = pd.DataFrame()
for i in range(5):
    title = 'y'+str(i)+'_title'
    annual_headers = ['name']
    for col in company_data.columns:
        if col[:8] == title or col[:9]==title:
            annual_headers.append(col)
    #all account header cols found for one year
    all_accounts = transpose_append_annual(sub_company_data,annual_headers,all_accounts)
print(all_accounts.info())
#print(all_accounts)
```

```python
all_accs = (all_accounts['Get Invited'].dropna(how='all'))


all_df = pd.DataFrame(all_accs)
company_name = all_df.columns.values[0]
print(company_name)
all_df.columns = ['acc_values']
all_df['year_title'] = all_df.index.str.split('_')
all_df['year'] = all_df['year_title'].str.get(0)
all_df['year'] = all_df['year'].str[1:]
all_df['title'] = all_df['year_title'].str.get(2)
#all_df['title'] = all_df['title'].str[:10] # this line was just to make reading a wee bit easier
all_df=all_df.drop(columns=['year_title'])
all_df.reset_index(drop=True, inplace=True) 
#print(all_df.info())
#print(all_df)

progress_df = all_df.pivot(index='title',columns='year', values='acc_values')

print(progress_df.info())
#print(progress_df)
for col in progress_df.columns:
    print(progress_df[col])
    #progress_df[col] = progress_df[col].str.strip()
    #progress_df[col] = progress_df[col].str.replace('-','')
    #progress_df[col] = progress_df[col].str.replace('(','-')
    #progress_df[col] = progress_df[col].str.replace(')','')
    #print(progress_df[col])
    progress_df[col] = pd.to_numeric(progress_df[col],downcast='signed', errors='coerce')
    
    print(progress_df[col])
    print("***** ***** **** ****")
```

```python
years = progress_df.columns.values
#progress_df=progress_df.dropna(how='any')
headings = progress_df.index.values
for heading in headings:
    print(heading)
    annual_vals = progress_df.loc[progress_df.index==heading].values[0] 
    plt.plot(years,annual_vals)
    plt.show()
    
```
